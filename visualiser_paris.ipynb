# 📊 Visualisation des Value Bets du Jour

```
import pandas as pd
import plotly.express as px
```

```
bets = pd.read_csv('data/bets_today.csv')
bets.sort_values('expected_value', ascending=False, inplace=True)
bets.head()
```

```
px.bar(bets, x='match', y='expected_value', color='bet_on', title='Value bets détectés')
```
